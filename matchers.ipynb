```python
import pandas as pd
import spacy
from spacy.matcher import PhraseMatcher
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_excel("hackdata.xlsx")
```

```python
abstracts = list(df['Abstract'].fillna('').astype(str))
processed_abstracts = [nlp(abstract) for abstract in abstracts]
```

```python
matcher = PhraseMatcher(nlp.vocab)
```

```python
search_query = nlp("drone")
matcher.add("SEARCH_QUERY", None, search_query)
```

```python
def get_data(query, abstracts):
    matches = []
    for index, doc in enumerate(abstracts):
        matches += [(index, start, end) for _, start, end, in matcher(doc)]
    return matches
```

```python
matched_abstracts = get_data(search_query, processed_abstracts)
```

```python
for index, start, end in matched_abstracts:
    print(f"Title: {df.iloc[index]['Title']}")
    print(f"Abstract: {df.iloc[index]['Abstract'][start:end]}")
```
